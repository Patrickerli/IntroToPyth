A veterinarian with the phone number 738-2273 would be pleased to know that the number corresponds to the letters “PETCARE.”  Write a script that, given a seven-digit number, generates every possible seven-letter word combination corresponding to that number. There are 2,187 (37) such combinations. Avoid phone numbers with the digits 0 and 1 (to which no letters correspond). See if your phone number corresponds to meaningful words.

```python
"""
1.  Get phone number from user as input
2.  Check if input is 7 digits long
3.  If not, prompt user to enter a valid 7-digit phone number
4.  Create a list of lists containing letters associated with phone keypad digits
5.  Iterate through the digits of the phone number
6.  Append the corresponding list of letters to a new list
7.  Generate all possible combinations of letters in the new list
8.  Append the combinations to a final list
9.  Print the final list of all possible "words" formed from the phone number
"""


# Prompt user to enter a 7-digit phone number
phone_number = input("please enter 7digits phone number: ")

# Check if input is 7 digits long
length = len(phone_number)

while length != 7 or "1" in phone_number or "0" in phone_number:
    phone_number = input("please enter 7digits phone number: ")
    length = len(phone_number)

# Create a list of lists containing letters associated with phone keypad digits
a = ["A","B","C"]#2
b = ["D","E","F"]#3
c = ["G","H","I"]#4
d = ["J","K","L"]#5
e = ["M","N","O"]#6
f = ["P","R","S"]#7
g = ["T","U","V"]#8
h = ["W","X","Y"]#9
alphabet = [a,b,c,d,e,f,g,h]

# Initialize empty list
collection = []

# Iterate through the digits of the phone number
# Append the corresponding list of letters to a new list
for p in range(7):
        for q in range(2,10):
                if int(phone_number[p]) == int(q):
                    collection.append(alphabet[q-2])

# Initialize empty list
collection_end = []

# Generate all possible combinations of letters in the new list
# Append the combinations to a final list
for i in collection[0]:
    for j in collection[1]:
        for k in collection[2]:
            for l in collection[3]:
                for m in collection[4]:
                    for n in collection[5]:
                        for o in collection[6]:
                            collection_end.append(i + j + k + l + m + n + o)

# Print the final list of all possible "words" formed from the phone number
print(collection_end, end = "")
```

```python
def product(*args, repeat=1):
    pools = [tuple(pool) for pool in args] * repeat
    result = [[]]
    for pool in pools:
        result = [x+[y] for x in result for y in pool]
    for prod in result:
        yield tuple(prod)

phone_number = input("please enter 7digits phone number: ")
length = len(phone_number)
while length != 7 or "1" in phone_number or "0" in phone_number:
    phone_number = input("please enter 7digits phone number: ")
    length = len(phone_number)

a = ["A","B","C"]#2
b = ["D","E","F"]#3
c = ["G","H","I"]#4
d = ["J","K","L"]#5
e = ["M","N","O"]#6
f = ["P","R","S"]#7
g = ["T","U","V"]#8
h = ["W","X","Y"]#9
alphabet = [a,b,c,d,e,f,g,h]
collection = []
for i in range(7):
        for j in range(2,10):
                if int(phone_number[i]) == int(j):
                    collection.append(alphabet[j-2])


letter_combinations = list(product(*collection))

for combination in letter_combinations:
    print("".join(combination), end = " ")          
```
