```python
"""
(Palindromes) A palindrome is a number, word or text phrase that reads the same
backwards or forwards. For example, each of the following five-digit integers is a palindrome: 12321, 55555, 45554 and 11611. Write a script that reads in a five-digit integer
and determines whether it’s a palindrome. [Hint: Use the // and % operators to separate
the number into its digits.]
"""
```

```python
#method 1
# Check if the number is a palindrome
"""
START

INPUT number

SET digit1 to number // 10000
SET digit2 to (number % 10000) // 1000
SET digit3 to (number % 1000) // 100
SET digit4 to (number % 100) // 10
SET digit5 to number % 10

IF digit1 equals digit5 AND digit2 equals digit4 THEN
  PRINT "The number is a palindrome."
  
ELSE
  PRINT "The number is not a palindrome."

END
"""

# Read in the integer
number = int(input("Enter a five-digit integer: ")) 

# Separate the digits of the integer
digit1 = number // 10000
digit2 = (number % 10000) // 1000
digit3 = (number % 1000) // 100
digit4 = (number % 100) // 10
digit5 = number % 10

# Check if the number is a palindrome
if digit1 == digit5 and digit2 == digit4:
    print("The number",number, "is a palindrome.")
else:
    print("The number",number, "is not a palindrome.")
```

```python
#method 2
# Check if the number is a palindrome
"""
START

INPUT a

SET x to string version of a

SET flag to True

FOR i in range 0 to length of x divided by 2:
  
  IF x at position i does not equal x at position -i - 1 THEN
    
    SET flag to False
    
    BREAK


IF flag is True THEN
  PRINT a, "is a palindrome!"
ELSE
  PRINT a, "is not a palindrome!"

END
"""

# Read in the integer
# # Convert the integer to a string
# # Set the flag to True
a = int(input("Please enter a number:\n"))  
x = str(a)  
flag = True 
 
 # Iterate over half of the characters in the string
 # Check if the characters at the current and opposite positions are different
 # Set the flag to False
 # Break out of the loop
for i in range(len(x)//2):  
    if x[i] != x[-i - 1]:   
        flag = False    
        break   

# Check the value of the flag
if flag:    
    print ("The number", a, "is a palindrome!")
else:
    print ("The number", a, "is not a palindrome!")
```

Drivers are concerned with the mileage obtained by their automobiles.
One driver has kept track of several tankfuls of gasoline by recording miles driven
and gallons used for each tankful. Develop a sentinel-controlled-repetition script that
prompts the user to input the miles driven and gallons used for each tankful. The script
should calculate and display the miles per gallon obtained for each tankful. After processing
all input information, the script should calculate and display the combined miles per
gallon obtained for all tankfuls (that is, total miles driven divided by total gallons used).

```python
""" Requirements

Input the amount of fuel used and mileage driven.
Outputs separate gallons per mile and the summed gallons per mile.

Requirements """



""" Pseduocode

SET total_gallons = 0
SET total_miles = 0
INPUT gallons
WHILE gallons ≠ -1
    INPUT miles
    OUTPUT miles / gallons
    SET total_gallons = total_gallons + gallons
    SET total_miles = total_miles + miles
    INPUT gallons
    IF gallons = -1
        IF total_gallons ≠ 0
            SET average = total_miles / total_gallons
            OUTPUT average
            END
ELSE
OUTPUT "No gallons were entered"

Pseduocode """



# initialization phase
total_gallons = 0  # sum of gallons
total_miles = 0  # sum of miles

# processing phase
gallons = float(input('Enter the gallons used (-1 to end): '))  # get gallons
while gallons != -1:
    miles = float(input('Enter the miles driven: '))  # get gallons
    print("The miles/gallon for this tank was",miles / gallons)
    
    total_gallons += gallons
    total_miles += miles

    gallons = float(input('Enter the gallons used (-1 to end): '))
    if gallons == -1:
    # termination phase
        if total_gallons != 0:
            average = total_miles / total_gallons
            print('The overall average miles/gallon was',average)
        break
else:
    print('No gallons were entered')
```

```python
import random

def tortoise_run():
    percentage_number = random.randint(1, 10)
    if percentage_number <= 5:
        return 3 
    elif percentage_number <= 7:
        return -6
    else:
        return 1

def hare_run():
    percentage_number = random.randint(1, 10)
    if percentage_number <= 2:
        return 0 
    elif percentage_number <=4:
        return 9
    elif percentage_number <=5:
        return -12
    elif percentage_number <=8:
        return 1
    else:
        return -2

tortoise_run()
hare_run()

tortoise_pos = int(1)
hare_pos = int(1)
        
print("BANG !!!!!\nAND THEY'RE OFF !!!!!\n")  
race_distance = ["_"]*70

while True:
    tortoise_pos += int(tortoise_run())
    hare_pos += int(hare_run())

    if tortoise_pos >= 70 or hare_pos >= 70:
        if tortoise_pos > hare_pos:
            print("TORTOISE WINS!!! YAY!!!")
        elif hare_pos > tortoise_pos:
            print("Hare wins. Yuch.")
        else:
            print("It's a tie")
        break

    tortoise_pos = max(1, tortoise_pos)
    hare_pos = max(1, hare_pos)
    race_distance[tortoise_pos-1] = 'T'
    race_distance[hare_pos-1] = 'H'

    if tortoise_pos == hare_pos:
        race_distance[tortoise_pos-1] = 'O'
        print("".join(race_distance),"OUCH!!!")

    else:
        print("".join(race_distance))
    race_distance = ['_']*70
```
