Try again." as appropriate to help the player “zero in” on the correct answer, then prompt the user for the next guess. When the user enters the correct answer, display "Congratulations. In problems 4.10 and 4.11, this is less clear -- but you should have the following functions in each:

```python
"""
# Guess the number game
IMPORT random

# Function to get a random number
FUNCTION get_number():
    RETURN random.randint(1, 1000)
END FUNCTION

# Function to check if the guess is correct
FUNCTION check_guess(guess, number):
    IF guess > number:
        PRINT "Too high. Try again."
    ELSEIF guess < number:
        PRINT "Too low. Try again."
    ELSE:
        PRINT "Congratulations. You guessed the number!"
        RETURN true
    RETURN false
END FUNCTION

# Function for a new game
FUNCTION new_game():
    number = get_number()
    PRINT "Guess my number between 1 and 1000 with the fewest guesses:"
    WHILE true
        INPUT guess
        IF check_guess(guess, number) = true
            BREAK
    END WHILE
END FUNCTION

WHILE true
    CALL new_game()
    INPUT "Do you want to play again?(y/n)"
    again = again.strip().lower()
    IF again not equals 'y'
        BREAK
END WHILE
"""
import random

def get_number():
    """Generates new random number to be guessed"""
    # choose the number to be guessed
    return random.randint(1, 1000)

def check_guess(guess, number):
    """Compare user's guess to the number the computer has chosen"""
    if guess > number:
        print("Too high. Try again.")
    elif guess < number:
        print("Too low. Try again.")
    else:
        print("Congratulations. You guessed the number!")
        return True
    return False

def new_game():
    """Starts a new game by generating a new number, displaying a prompt, and repeatedly prompting for guesses"""
    number = get_number()
    print("Guess my number between 1 and 1000 with the fewest guesses:")
    while True:
        guess = int(input())
        if check_guess(guess, number):
            break

# Keeps looping the new_game function until user decides to stop playing
while True:
    new_game()
    again = input("Do you want to play again?(y/n)").strip().lower()
    if again != 'y':
        break
```
