Read the stopword file

Initialize a variable to keep track of removed stopwords

Check if the word is a stopword, if it is, increase the count of removed stopwords and move on to the next word

```python
def main():
    # This program analyzes word frequency in a file
    print("This program analyzes word frequency in a file")
    print("and prints a report on the n most frequent words.\n")

    # Take the file name as input
    file_name = input("Please enter the file you want:")
    # Read the file
    text = open(file_name, "r").read()

    # Initialize an empty dictionary to store word frequency
    word_counts = {}

    # Convert the text to lowercase
    text = text.lower()

    # Replace all punctuation marks with a space
    for i in text:
        if i in '!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~':
            text = text.replace(i, " ")

    # Take the stopword file name as input
    stopword_name = input("Please enter the stopword_file :")
    # Read the stopword file
    stopfile = open(stopword_name, "r").read()
    stopword = stopfile.split()

    # Initialize variable to keep track of removed stopwords
    removeword = 0
    # Iterate through each word in the text
    for word in text.split():
        # Check if the word is a stopword
        while word in stopword:
            # Increase the count of removed stopwords
            removeword += 1
            break
        # If the word is not a stopword
        while word not in stopword:
            # Check if the word already exists in the word_counts dictionary
            if word in word_counts:
                # If it does, increase the count by 1
                word_counts[word] += 1
                break
            else:
                # If it doesn't, add the word to the dictionary with a count of 1
                word_counts[word] = 1
                break

    # Split the text into a list of words
    words = text.split(" ")
    # Convert all words to lowercase
    words = [word.lower() for word in words]

    # Get a set of unique words
    unique_words = set(words)

    # Sort the unique words
    unique_words = sorted(unique_words)

    # Sort the word frequency dictionary by frequency
    listnumber = sorted(word_counts.items(), key = lambda kv:(kv[1], kv[0]), reverse= True)
    # Take the number of words to output as input
    word_counts_input = int(input("Output analysis of words:"))
    # Print the file name and the number of words to output
    print("\nFile to analyze:",file_name)
    print("\nOutput analysis of words:", word_counts_input)
    print("\nRemoveword of words:",removeword)
    print()
    # Print the header for the word frequency table
    print(f'{"WORD":<15}COUNT')

    # Initialize a count variable
    count_number = 0
    # Iterate through the sorted word frequency list
    for word, count in listnumber:
        count_number += 1
        # Print the word and its frequency until the specified number of words is reached
        if count_number <= word_counts_input:
            print(f"{word:<15}{count}")
            
        else: 
            break

main()
```
